**1. Line Plot â†’ plt.plot()**

ðŸ‘‰ Used to show trends over time or continuous data.

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[5,6,7,8]
plt.plot(x,y,marker='o',color='k',label="y=2x")
plt.title("Line Plot Example")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

***2. Bar Chart â†’ plt.bar(), Horizontal â†’ plt.barh()***

ðŸ‘‰ Used to compare categorical data.

```python
fruits=[ "lemon","mango","watermelon"]
values=[10,15,20]
plt.bar(fruits,values,color="green")
plt.title("Bar Chart Example")
plt.show()

# Horizontal bar chart
plt.barh(fruits, values, color="orange")
plt.title("Horizontal Bar Chart Example")
plt.show()
```

3. Scatter Plot â†’ plt.scatter()

ðŸ‘‰ Used to show relationships/correlations between two variables.

```python
import numpy as np

# Random data
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color="red", marker="*")
plt.title("Scatter Plot Example")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.show()
```

4. Histogram â†’ plt.hist()

ðŸ‘‰ Used to show the distribution of data (frequency of values).

```python
data = np.random.randn(1000)

plt.hist(data, bins=20, color="purple", edgecolor="black")
plt.title("Histogram Example")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```

5. Pie Chart â†’ plt.pie()

ðŸ‘‰ Used to show parts of a whole (percentages).

```python
sizes=[15,25,35,40,45]
labels=["A","B","C","D","E"]
colors=["red","blue","green","orange","black"]
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%",startangle=90)
plt.title("pie chart example")
plt.show()
```

6. Stacked Plots (Multiple datasets in one chart)

ðŸ‘‰ Useful for showing comparison of multiple series in one plot.

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 4, 8, 16]
y2 = [1, 3, 6, 10, 15]

plt.plot(x, y1, label="Series 1")
plt.plot(x, y2, label="Series 2")
plt.title("Multiple Line Plot Example")
plt.legend()
plt.show()
```

(b) Stacked Bar Plot

```python
import numpy as np

categories = ["A", "B", "C"]
values1 = [3, 5, 7]
values2 = [2, 4, 6]

plt.bar(categories, values1, label="Group 1")
plt.bar(categories, values2, bottom=values1, label="Group 2")  # stacked
plt.title("Stacked Bar Chart Example")
plt.legend()
plt.show()
```

**5. Object-Oriented Interface**

ðŸ“Œ Milestone: Build a dashboard-like figure with 4â€“6 subplots.

**1. fig, ax = plt.subplots()**:

ðŸ‘‰ Creates a Figure (fig) and an Axes (ax) object.

Figure = entire canvas

Axes = single plot area

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots()
x=[1,2,3,4]
y=[5,6,7,8]
ax.plot(x, y, marker="o", color="blue")
ax.set_title("fig,ax line plot")
ax.set_xlabel("x values")
ax.set_ylabel("y values")
plt.show()
```

2. Multiple plots with same Axes â†’ ax.plot(), ax.bar(), etc.

ðŸ‘‰ You can call different methods on the same ax.

```python
fig,ax=plt.subplots()
x=[1,2,3,4]
y1=[5,6,7,8]
y2=[9,10,11,12]
ax.plot(x,y1,label="y1=x^2",color="red")
ax.bar(x,y2,alpha=0.5,label="bar values")
ax.set_title("multiple plots")
ax.legend()
plt.show()
```

 3. Multiple Subplots â†’ plt.subplots(nrows, ncols)

ðŸ‘‰ Arrange multiple Axes in a grid layout.

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y=np.sin(x)
fig,axes=plt.subplots(2,2,figsize=(8,6))
axes[0, 0].plot(x, y)
axes[0, 0].set_title("Sine Wave")
axes[0,1].scatter(np.random.rand(50),np.random.rand(50),color="red")
axes[0,1].set_title("scatter plot")
axes[1,0].bar(["A","B","C"],[3,5,7],color="orange")
axes[1,0].set_title("bar chart")
axes[1,1].hist(np.random.rand(200),bins=20,color="green")
axes[1,1].set_title("hist")
plt.tight_layout()
plt.show()


```

4. Adjusting Layouts â†’ plt.tight_layout()

ðŸ‘‰ Prevents overlapping titles/labels when you have many subplots.

Example above already used it.
Without plt.tight_layout(), the axis labels/titles might overlap.

ðŸ“Œ Milestone: Dashboard with 4â€“6 Subplots

```python
fig,axes=plt.subplots(2,3,figsize=(12,6))
axes[0,0].plot(x,y,color="blue")
axes[0,0].set_title("line plot")
axes[0,1].scatter(np.random.rand(50),np.random.rand(50),color="black")
axes[0,1].set_title("scatter plot")
axes[0,2].bar(["A","B","C"],[3,5,7],color="pink")
axes[0,2].set_title("bar chart")
axes[1, 0].hist(np.random.randn(100), bins=10, color="purple")
axes[1, 0].set_title("Histogram")
axes[1, 1].pie([30,20,50], labels=["X","Y","Z"], autopct="%.1f%%")
axes[1, 1].set_title("Pie Chart")
axes[1, 2].plot(x, np.sin(x), label="sin")
axes[1, 2].plot(x, np.cos(x), label="cos")
axes[1, 2].legend()
axes[1, 2].set_title("sin & cos")

plt.tight_layout()
plt.show()
```

***âœ… Output: A dashboard-like figure with 6 different plots in one window.***

ðŸ“Œ Milestone: Make a scientific plot with error bars, labels, and annotations.

***1. Annotations & Text â†’ ax.annotate(), ax.text()***

ðŸ‘‰ Add notes, arrows, or labels inside the chart.

***In Matplotlib, annotations & text let you put extra information (like labels, arrows, notes) directly inside your plot.***

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y=np.sin(x)
fig,ax=plt.subplots()
ax.plot(x,y,label="sin(x)")
ax.text(2,0.5,"peak here",fontsize=10,color="red")
ax.annotate("zero crossing",xy=(3.14,0),xytext=(4,0.5),arrowprops=dict(facecolor="black",arrowstyle="->"))
ax.set_title("annotation example")
ax.legend()
plt.show()
```

2. Error Bars â†’ plt.errorbar()

ðŸ‘‰ Show uncertainty or variability in data points.

```python
x = np.arange(1, 6)
y = [2, 4, 6, 8, 10]
errors = [0.5, 0.7, 0.3, 0.9, 0.4]
plt.errorbar(x,y,yerr=errors,fmt='o',capsize=5,color="blue")
plt.title("error bar")
plt.show()
```

3. Multiple Axes â†’ twinx(), Secondary Axes

ðŸ‘‰ Show two different scales in the same plot.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.exp(0.1 * x)

fig, ax1 = plt.subplots()

ax1.plot(x, y1, color="blue", label="sin(x)")
ax1.set_ylabel("sin(x)", color="blue")
ax2 = ax1.twinx()
ax2.plot(x, y2, color="red", label="exp(0.1x)")
ax2.set_ylabel("exp(0.1x)", color="red")
plt.title("Multiple Axes Example")
plt.show()
```

4. Logarithmic Scale â†’ ax.set_xscale("log")

ðŸ‘‰ Useful when data spans orders of magnitude.

```python
x = np.logspace(0.1, 2, 100)  # 10^0.1 to 10^2
y = x**2

fig, ax = plt.subplots()
ax.plot(x, y, color="green")
ax.set_xscale("log")  # log scale on x-axis
ax.set_yscale("log")  # log scale on y-axis
ax.set_title("Logarithmic Scale Example")
plt.show()
```

5. Custom Ticks & Formatting â†’ FuncFormatter

ðŸ‘‰ Control how ticks look (e.g., percentages, currency)

```python
from matplotlib.ticker import FuncFormatter
x = [1, 2, 3, 4, 5]
y = [0.1, 0.25, 0.5, 0.75, 1.0]
fig, ax = plt.subplots()
ax.plot(x, y, marker="o")
formatter=FuncFormatter(lambda val,pos:f"{val*100:.0f}%")
ax.yaxis.set_major_formatter(formatter)
ax.set_title("Custom Tick Formatting Example")
plt.show()
```

*** Styling & Themes:***
1. Stylesheets â†’ plt.style.use()

ðŸ‘‰ Matplotlib comes with built-in themes (similar to filters).
Popular: "ggplot", "seaborn", "fivethirtyeight", "dark_background".

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

# Apply ggplot style
plt.style.use("ggplot")
plt.plot(x, y, label="sin(x)")
plt.title("Style: ggplot")
plt.legend()
plt.show()

# Apply fivethirtyeight style
plt.style.use("fivethirtyeight")
plt.plot(x, y, label="sin(x)")
plt.title("Style: fivethirtyeight")
plt.legend()
plt.show()
```

**all styles at once:**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

styles = ["ggplot", "seaborn-v0_8-bright", "fivethirtyeight", "dark_background"]

for style in styles:
    plt.style.use(style)           # Apply the style
    plt.plot(x, y, label="sin(x)")
    plt.title(f"Style: {style}")   # Show which style is used
    plt.legend()
    plt.show()
```

2. Customizing rcParams â†’ plt.rcParams

ðŸ‘‰ rcParams = global configuration dictionary.
You can control figure size, font size, line width, etc.

```python
plt.rcParams["figure.figsize"]=(8,6)
plt.rcParams["font.size"]=12
plt.rcParams["lines.linewidth"]=3
x = np.linspace(0, 5, 50)
y = np.exp(x)
plt.plot(x, y, color="purple", label="exp(x)")
plt.title("Custom rcParams Example")
plt.legend()
plt.show()
```

3. Colormaps â†’ plt.cm + cmap= argument

ðŸ‘‰ Useful for gradients or mapping values to colors (heatmaps, scatter plots).

```python
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
plt.scatter(x,y,c=colors,cmap=plt.cm.viridis,s=80)
plt.colorbar(label="color scale")
plt.title("Colormap Example (viridis)")
plt.show()
```

4. Saving Figures â†’ plt.savefig()

ðŸ‘‰ Save plots in different formats (PNG, PDF, SVG) with high resolution.

```python
x = np.linspace(0, 10, 100)
y = np.cos(x)

plt.plot(x, y, color="teal", label="cos(x)")
plt.title("Saving Figures Example")
plt.legend()

# Save as PNG (300 dpi = high quality for publications)
plt.savefig("cosine_plot.png", dpi=300)

plt.show()
```

***Advanced Plots:***

***1. Heatmaps â†’ imshow()***

ðŸ‘‰ Represent 2D data as a color grid.

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(6, 6)
plt.imshow(data,cmap="viridis",interpolation="nearest")
plt.colorbar(label="Value")
plt.title("Heatmap Example")
plt.show()
```

2. Contour Plots â†’ contour(), contourf()

ðŸ‘‰ Show curves of equal value in 2D functions (like topographic maps)

```python
x=np.linspace(-3,3,100)
y=np.linspace(-3,3,100)
x,y=np.meshgrid(x,y)
z=np.sin(np.sqrt(x**2 +y**2))
plt.contour(x,y,z,levels=10,color="blue")     #levels=10 â†’ divide Z-values into 10 equally spaced levels
plt.contourf(x,y,z,levels=10,cmap="viridis")
plt.colorbar(label="vaues")
plt.title("contour plot")
plt.show()

```

Integration with Pandas

1. Line Plots â†’ df.plot(kind="line")

ðŸ‘‰ Best for trends across time/index.

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Sales": [200, 250, 300, 280, 320]
}
df = pd.DataFrame(data)
df.set_index("Month", inplace=True)
df.plot(kind="line", marker="o", title="Monthly Sales")
plt.ylabel("Sales Amount")
plt.show()
```

2. Overlays with Multiple Columns

ðŸ‘‰ Plot multiple columns on the same chart.

```python
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Sales": [200, 250, 300, 280, 320],
    "Expenses": [150, 180, 210, 200, 230]
}
df = pd.DataFrame(data)
df.set_index("Month", inplace=True)
df.plot(kind="line", marker="o", title="Sales vs Expenses")
plt.ylabel("Amount")
plt.show()
```

3. Time Series Plots

ðŸ‘‰ Pandas automatically handles dates as x-axis

```python
import numpy as np
dates=pd.date_range("2023-01-01",periods=10)
df = pd.DataFrame({
    "Date": dates,
    "Temperature": np.random.randint(20, 35, size=10)
})
df.set_index("Date",inplace=True)
df.plot(title="Daily Temperature")
plt.ylabel("Temperature (Â°C)")
plt.show()
```

4. Histogram from DataFrame

ðŸ‘‰ Show distribution of values from a column.

```python
df["Temperature"].plot(kind="hist",bins=5,color="pink",edgecolor="green")
plt.title("Temperature Distribution")
plt.xlabel("Temperature (Â°C)")
plt.show()
```
