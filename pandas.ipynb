**Pandas:**

Definition: Pandas is an open-source Python library used for data manipulation and analysis.

Built on: Numpy (provides fast numerical computation).

Main Data Structures:

Series (1D data)

DataFrame (2D data)

Key Features:

Handling missing data

Easy data filtering, grouping, and aggregation

Supports CSV, Excel, JSON, SQL, Parquet file formats

Time series support

**Series:**

Definition: A one-dimensional labeled array in Pandas.

Can store: integers, floats, strings, Python objects, etc.

Like: A single column in Excel or a 1D array with labels (index).

**Features:**

Indexing & slicing like arrays

Vectorized operations (fast computations)

Alignment based on index.


**DataFrame:**

Definition: A two-dimensional labeled data structure (rows & columns).

Like: Excel spreadsheet or SQL table.

Stores: multiple Series as columns.

**Features:**

Supports heterogeneous data (different data types per column)

Easy row/column selection (df["Name"], df.loc[0])

Data cleaning, merging, grouping, and summarizing

Handles large datasets efficiently.

***grouping data:***

```python
# Step 1: create the column
df['Workout_Type'] = df['Duration'].apply(
    lambda x: 'Short' if x < 45 else ('Medium' if x == 45 else 'Long')
)

# Step 2: now group by it
workout_groups = df.groupby('Workout_Type')

# Step 3: try some aggregations
print(workout_groups['Calories'].mean())   # avg Calories by type
print(workout_groups['Pulse'].max())       # max Pulse by type
```

***Aggregation:After grouping we can perform aggregation operations like sum, count,mean etc..***

```python
df.groupby('Duration').mean()
df
```

```python
#different aggregations on different coumns
summary = df.groupby('Duration').agg({
    'Pulse': ['mean', 'std', 'min', 'max'],
    'Maxpulse': ['mean', 'std', 'min', 'max'],
    'Calories': ['mean', 'std', 'min', 'max']
})

print(summary)
```

```python
#multiple aggregations:# group by Duration and apply multiple aggregations
# group by Duration, aggregate only Calories
summary = df.groupby('Duration')['Calories'].agg(['mean', 'std', 'min', 'max'])
print(summary)
```

***custom aggreegate func:***

```python
#Example 1: Custom function for range (max  min)
def value_range(x):
    return x.max() - x.min()

summary = df.groupby('Duration')['Calories'].agg(['mean', 'max', 'min', value_range])
print(summary)
```

```python
#Example 2: Lambda function directly
summary = df.groupby('Duration')['Pulse'].agg([
    'mean',
    lambda x: (x > 110).sum()   # count how many pulses > 110
])
summary.columns = ['mean', 'count_above_110']
print(summary)
```

```python
#Multiple custom + built-in functions
def my_custom(x):
    return round((x.sum() / len(x)), 2)  # custom mean

summary = df.groupby('Duration')['Calories'].agg(['mean', my_custom])
print(summary)
```

***renaming resuts of aggregation for clarity.***

```python
def my_custom(x):
    return round((x.sum() / len(x)), 2)  # custom mean

summary = (
    df.groupby('Duration')['Calories']
      .agg(Calories_mean='mean', Calories_custom=my_custom)
)

print(summary)
```
