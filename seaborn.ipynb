Understanding Seaborn vs Matplotlib

Built-in Datasets (sns.get_dataset_names(), sns.load_dataset())

Seaborn Themes & Styles (set_theme, set_style, set_palette)

Basic Plot Structure (sns.function(data=..., x=..., y=...))

***Why Matplotlib is used***

Matplotlib is the foundation of Python data visualization.

It’s like the "low-level painter" — you control every detail.

It provides 2D plotting and can be extended to 3D.

**✅ Uses of Matplotlib:**

Basic plotting – line, bar, scatter, histogram, pie.

Fine-grained control – axis ticks, labels, legends, styles.

Custom plots – since you can build any type of graph manually.

Integration – works well with NumPy, Pandas.

Publication-quality figures – used for scientific research, reports.

***Why Seaborn is used:***

Seaborn is built on top of Matplotlib.

It’s like the "interior designer" — it makes plots look beautiful and easy.

Focuses on statistical visualization with fewer lines of code.

***✅ Uses of Seaborn:***

Attractive default styles – better than raw Matplotlib.

Statistical plots – boxplot, violinplot, heatmap, pairplot.

Works directly with Pandas DataFrames – no need to manually extract columns.

Complex plots in one line – distribution, regression, correlation.

| Feature             | Matplotlib                 | Seaborn                                 |
| ------------------- | -------------------------- | --------------------------------------- |
| **Level**           | Low-level (more control)   | High-level (simpler)                    |
| **Look & Style**    | Plain by default           | Stylish by default                      |
| **Ease of Use**     | More code needed           | Less code needed                        |
| **Customizability** | Very high                  | Moderate (but still good)               |
| **Best for**        | Custom plots, fine control | Quick statistical plots, pretty visuals |


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Matplotlib Example
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, color='blue')
plt.title("Matplotlib Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

sns.lineplot(x=x,y=y)
plt.title("Seaborn Plot")
plt.show()

```

1. Bivariate Distributions
✅ Definition:

Bivariate distribution means the relationship between two variables (x and y).

In Seaborn, we can visualize them using scatterplots, lineplots, and jointplots.

Helps in spotting correlations, trends, or clusters.

2. sns.scatterplot()
✅ Definition:

A scatter plot shows individual data points for two variables.

Each dot represents an observation (x, y).

We can use hue, style, or size to add extra information.

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="smoker",style="sex")
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()
```

3. sns.lineplot()
✅ Definition:

A line plot connects data points to show trends over continuous variables.

Often used for time series or continuous numeric relationships.

By default, Seaborn calculates the mean for y-values if data is categorical on x.

```python
# Line plot: average tip per day
sns.lineplot(data=tips, x="day", y="tip", hue="sex", marker="o")
plt.title("Line Plot: Average Tip per Day (by Sex)")
plt.show()
```

Example 2 (Continuous Data):

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
x=np.linspace(0,10,100)
y=np.sin(x)+np.random.normal(scale=0.2,size=100)
df = pd.DataFrame({"x": x, "y": y})
sns.lineplot(data=df, x="x", y="y")
plt.title("Line Plot with Continuous Data")
plt.show()
```

4. sns.jointplot()
✅ Definition:

Combines a scatterplot with marginal distributions (histograms or KDEs).

Helps to see both the relationship AND the distribution of variables.

Can use kind="scatter", kind="reg", kind="hex", or kind="kde".

```python
# Jointplot scatter + hist
sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter", hue="sex")
plt.show()
```

Output (visual explanation):

Center: Scatter plot of total_bill vs tip.

Top: Histogram of total_bill.

Right: Histogram of tip.
👉 Shows correlation + distribution of both variables.

What is Regression?

Regression is a statistical method to find the relationship between variables.

It fits a line or curve that best describes how one variable (X) affects another variable (Y).

In Seaborn, regression is mostly linear regression (straight line fit).

Example 2: With Regression Line

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="reg")
plt.show()
```

Output:

Scatter plot with a regression line (best-fit line).

Marginal histograms show the distribution.
👉 Quickly shows correlation + trend.

What is KDE?

KDE = Kernel Density Estimation

It is a way to estimate the probability distribution of a dataset smoothly, instead of just using histograms.

Think of it as a smoothed histogram that shows where data points are concentrated.

example:3 KDE (Density)

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="kde", fill=True)
plt.show()
```

Output:

Contour plot (like a heatmap) showing density of points.

Darker regions = more data points.
👉 Good for large datasets (avoids overplotting).

Summary

sns.scatterplot() → Shows raw data points (best for correlations).

sns.lineplot() → Shows trends (good for time series or continuous data).

sns.jointplot() → Combines scatter + distribution (hist, KDE, or regression).

Categorical plots are used when one variable is categorical (like gender, day, smoker/non-smoker, etc.) and the other is numerical or also categorical.

We’ll cover barplot, countplot, boxplot, violinplot, stripplot, and swarmplot

1. sns.barplot()
✅ Definition:

Shows the average of a numerical variable for each category.

Displays confidence intervals (by default, 95% CI).

Good for comparing group means.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Barplot: average tip per day
sns.barplot(data=tips, x="day", y="tip", hue="sex")
plt.title("Bar Plot: Average Tip by Day & Gender")
plt.show()
```

Output (visual explanation):

X-axis = Days (Thur, Fri, Sat, Sun).

Y-axis = Average tip amount.

Bars are grouped by Male/Female (hue).
👉 Shows which gender tips more on each day

2. sns.countplot()
✅ Definition:

Shows the count of observations per category.

It is basically like barplot but works with counts (frequencies).

```python
sns.countplot(data=tips, x="day", hue="sex")
plt.title("Count Plot: Number of Customers per Day")
plt.show()
```

Output:

Bars represent number of customers visiting on each day.

Hue = Male vs Female breakdown.
👉 Useful for categorical distributions.

3. sns.boxplot()
✅ Definition:

Also called a whisker plot.

Shows distribution of a numerical variable across categories.

Displays median, quartiles, outliers, and spread.

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Box Plot: Distribution of Bills by Day & Gender")
plt.show()
```

Output:

Each box shows:

Middle line = Median.

Box edges = Q1 (25%) and Q3 (75%).

Whiskers = spread (1.5 * IQR).

Dots = outliers.
👉 Great for spotting outliers and skewness.

4. sns.violinplot()
✅ Definition:

Combines boxplot + KDE (kernel density estimate).

Shows distribution shape (thick = many data points).

```python
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex",split=True)
plt.title("Violin Plot: Distribution of Bills by Day & Gender")
plt.show()
```

Output:

Violin shape indicates distribution.

Split violins (Male vs Female) side by side.
👉 Useful for seeing density + distribution.

5. sns.stripplot()
✅ Definition:

Shows individual data points (jittered for visibility).

Good for small datasets.

```python
sns.stripplot(data=tips, x="day", y="total_bill", jitter=True, hue="sex", dodge=True)
plt.title("Strip Plot: Individual Bills by Day")
plt.show()
```

Output:

Each dot = one bill.

Jitter spreads dots so they don’t overlap.
👉 Good for raw data visualization.

6. sns.swarmplot()
✅ Definition:

Similar to stripplot, but prevents overlapping points.

Points are arranged in a bee-swarm layout.

```python
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.title("Swarm Plot: Individual Bills by Day")
plt.show()
```

Output:

Each point shown without overlapping.

Looks like an organized scatterplot on categories.
👉 More readable than stripplot when dataset is not huge.

| Plot Type      | What it Shows               | When to Use            |
| -------------- | --------------------------- | ---------------------- |
| **barplot**    | Mean + CI per category      | Comparing averages     |
| **countplot**  | Count per category          | Frequency comparison   |
| **boxplot**    | Quartiles, median, outliers | Spread & outliers      |
| **violinplot** | Distribution + density      | Shape of distribution  |
| **stripplot**  | Raw data points (jittered)  | Small datasets         |
| **swarmplot**  | Raw points, no overlap      | Clearer than stripplot |

1. sns.heatmap()
✅ Definition:

A heatmap displays a matrix of values as a colored grid.

Each cell’s color intensity corresponds to the numeric value.

Often used for correlation matrices, confusion matrices, pivot tables, etc.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Compute correlation matrix
corr = tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm",linewidths=0.5)
plt.title("Correlation Heatmap of Tips Dataset")
plt.show()
```

Output (visual explanation):

Square grid showing correlations between columns (total_bill, tip, size, etc.).

Color from blue (-1 correlation) to red (+1 correlation).

Numbers inside cells are correlation coefficients.
👉 Helps detect strongly related variables (e.g., total_bill & tip are strongly correlated).

***What is a Pivot Table?***

A pivot table is a way to summarize and restructure your dataset:

You group data by some keys (like rows/columns).

You apply an aggregation function (like mean, sum, count, etc.).

You get a compact table that highlights patterns.

It’s like a multi-dimensional groupby with a nice table format.

Example 2: Pivot Table Heatmap

```python
pivot=tips.pivot_table(values="tip",index="day",columns="time",aggfunc="mean")
sns.heatmap(pivot, annot=True, cmap="YlGnBu")
plt.title("Average Tip by Day & Time")
plt.show()
```

Output:

Rows = Days, Columns = Time (Lunch/Dinner).

Each cell = mean tip value.

Colors show low vs high tips.
👉 Useful for summarization of grouped data.

***2. sns.clustermap()***

Definition:

A cluster map is a heatmap + hierarchical clustering.

It reorders rows and columns so that similar variables/observations are grouped together.

Useful for pattern detection in high-dimensional data.

Example 1: Correlation Clustermap

```python
# Clustermap with correlation matrix
sns.clustermap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Clustered Correlation Heatmap")
plt.show()
```

Output:

Similar to heatmap but adds dendrograms (tree-like structure) on top and left.

Rows/columns automatically reordered based on similarity.
👉 Groups strongly correlated variables together.

Example 2: Tips Data Matrix

```python
pivot2 = tips.pivot_table(values="total_bill", index="day", columns="sex", aggfunc="mean")
sns.clustermap(pivot2, annot=True, cmap="mako")
plt.title("Clustered Heatmap of Bills")
plt.show()
```

Output:

Heatmap of average bills by day vs sex.

Clustering may reorder days/sex to show similarity patterns.

| Feature    | `sns.heatmap()`           | `sns.clustermap()`              |
| ---------- | ------------------------- | ------------------------------- |
| Purpose    | Show values with color    | Show values + cluster rows/cols |
| Reordering | No (keeps order as given) | Yes (based on similarity)       |
| Dendrogram | No                        | Yes                             |
| Best for   | Quick summaries           | Pattern detection & grouping    |


Pairwise Plots

sns.pairplot()

sns.pairwise_kde()

Pairwise Plots in Seaborn, which are very powerful for exploring relationships between multiple variables.

1. sns.pairplot()

Definition:

A pair plot shows all pairwise relationships in a dataset.

It creates a grid of subplots:

Scatterplots for numeric vs numeric variables.

Histograms (or KDE plots) for diagonal (single variable distributions).

Often used in exploratory data analysis (EDA).

Example 1: Default Pairplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
iris = sns.load_dataset("iris")

# Pairplot
sns.pairplot(iris)
plt.show()
```

Output (visual explanation):

Scatterplots between each pair of numeric columns (sepal_length, sepal_width, petal_length, petal_width).

Histograms along diagonal.
👉 Helps see correlation (e.g., petal length & petal width strongly correlated).

Example 2: Pairplot with Hue

```python
sns.pairplot(iris, hue="species", diag_kind="kde")
plt.show()
```

Output:

Points are colored by species (Setosa, Versicolor, Virginica).

Diagonal uses KDE (smooth curves) instead of histograms.
👉 Helps separate clusters (Setosa clearly separated, others overlap).

2. sns.pairwise_kde()

Definition:

pairwise_kde() is a lower-level function used internally by pairplot.

It computes and plots the kernel density estimate (KDE) for pairwise combinations of variables.

KDE plots show density regions instead of individual points (useful for large datasets).

```python
# KDE for sepal_length vs sepal_width
sns.pairwise_kde(data=iris, vars=["sepal_length", "sepal_width"], fill=True)
plt.show()
```

Output:

Matrix of KDE plots for all combinations of the chosen variables.

Diagonal = 1D KDE curves.

Off-diagonal = 2D KDE density plots.
👉 Gives a smooth distribution overview.

*Summary: Pairplot vs Pairwise KDE:*

| Feature     | `sns.pairplot()`               | `sns.pairwise_kde()`                    |
| ----------- | ------------------------------ | --------------------------------------- |
| Type        | High-level plot                | Low-level function                      |
| Shows       | Scatterplots + Histograms/KDEs | Density plots (KDE only)                |
| Best for    | EDA (exploring all variables)  | Smooth density overview, large datasets |
| Hue support | Yes (categorical coloring)     | No direct hue support                   |
| Ease of use | Beginner-friendly              | Advanced use cases                      |


Regression & Trends

sns.regplot()

sns.lmplot()

Polynomial & logistic regression fits

1. sns.regplot()
✅ Definition:

regplot() plots a scatterplot plus a linear regression line.

It directly works with arrays or DataFrame columns.

By default, it fits a linear regression model (y = mx + c).

Example 1: Basic Regression

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Regression: total_bill vs tip
sns.regplot(data=tips, x="total_bill", y="tip")
plt.title("Regplot: Bill vs Tip")
plt.show()
```

Output:

Scatterplot of points.

Blue regression line with shaded confidence interval.
👉 Shows positive correlation: bigger bills → higher tips.

Example 2: Without Confidence Interval

```python
sns.regplot(data=tips, x="total_bill", y="tip", ci=None, color="red")
plt.title("Regplot Without CI")
plt.show()
```

Output:

Same regression line, but no shaded confidence band.

2. sns.lmplot()
✅ Definition:

Similar to regplot(), but more flexible:

Works with whole DataFrames.

Allows faceting by categories (row, col, hue).

Stands for Linear Model Plot.

Example 1: Basic lmplot

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.title("Lmplot: Regression by Gender")
plt.show()
```

Output:

Regression lines drawn separately for Male/Female.
👉 Useful for comparing trends across groups.

Example 2: Faceting by Rows & Columns

```python
sns.lmplot(data=tips,x="total_bill",y="tip",col="day",hue="sex",height=4)
plt.show()
```

Output:

Multiple subplots (one per day).

Each subplot shows regression lines for Male vs Female.
👉 Great for multi-panel comparisons.

3. Polynomial Regression Fits
✅ Definition:

Sometimes relationships are non-linear.

You can fit a polynomial regression by setting order in regplot/lmplot.

Example: quadratic (order=2), cubic (order=3)

Example: Quadratic Fit

```python
import numpy as np
import pandas as pd

# Generate data
x = np.linspace(-3, 3, 100)
y = x**2 + np.random.normal(0, 2, 100)
df = pd.DataFrame({"x": x, "y": y})

# Polynomial regression (quadratic)
sns.regplot(data=df, x="x", y="y", order=2, scatter_kws={"color":"blue"}, line_kws={"color":"red"})
plt.title("Polynomial Regression (Quadratic)")
plt.show()
```

Output:

Scatterplot shows parabola-shaped data.

Red regression line fits a quadratic curve.
👉 Useful when data isn’t linear.

4. Logistic Regression Fits
✅ Definition:

Logistic regression models a binary outcome (yes/no, success/failure).

In Seaborn, you can set logistic=True in regplot/lmplot.

Regression line will be an S-shaped curve (sigmoid).

Example: Logistic Fit

```python
tips["big_tip"]=(tips["tip"]>3) .astype(int)
sns.regplot(data=tips,x="total_bill",y="big_tip",logistic=True,scatter_kws={"color":"blue"})
plt.title("Logistic Regression: Probability of Big Tip")
plt.show()
```

Output:

Scatterplot with points at 0 or 1 (binary).

S-shaped curve shows probability of a "big tip" as total_bill increases.
👉 Helps in classification-type problems.

summary:

| Function            | Purpose                              | Best For               |
| ------------------- | ------------------------------------ | ---------------------- |
| **`sns.regplot()`** | Scatterplot + regression line        | Simple regression fits |
| **`sns.lmplot()`**  | Regression with faceting/hue         | Group comparisons      |
| **Polynomial Fit**  | Fits curves (quadratic, cubic, etc.) | Non-linear trends      |
| **Logistic Fit**    | Fits sigmoid curve                   | Binary classification  |

1. Color Palettes
✅ Definition:

Seaborn provides predefined color palettes and supports custom palettes for consistent, aesthetic plots.

Predefined palettes: "deep", "muted", "bright", "pastel", "dark", "colorblind"

Diverging palettes: "coolwarm", "RdBu", "Spectral", etc.

Custom colors: List of colors (["red", "blue", "green"])

Example 1: Predefined Palettes

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Predefined color palette
sns.set_palette("muted")
sns.barplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Bar Plot with Muted Palette")
plt.show()
```

Output:

Bars appear in soft, muted colors.

Example 2: Diverging Palette

```python
sns.set_palette("coolwarm")
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
plt.title("Scatter Plot with Coolwarm Palette")
plt.show()
```

Output:

Gradient-like colors from cool (blue) to warm (red).

Example 3: Custom Colors

```python
custom_colors = ["#FF5733", "#33FF57", "#3357FF"]
sns.set_palette(custom_colors)

sns.barplot(data=tips, x="day", y="tip", hue="sex")
plt.title("Custom Color Palette Example")
plt.show()
```

Output:

Bars colored with custom hex codes.

2. Custom Styles
✅ Definition:

Seaborn has plotting styles that change the background, grids, and ticks.

Options: "dark", "white", "whitegrid", "darkgrid", "ticks"

example

```python
sns.set_style("whitegrid")
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Box Plot with Whitegrid")
plt.show()
sns.set_style("dark")
sns.violinplot(data=tips, x="day", y="total_bill")
plt.title("Violin Plot with Dark Style")
plt.show()
```

Output:

First plot: clean white grid.

Second plot: dark background for dramatic effect.

3. Axis & Figure Control
✅ Definition:

You can control the axis (ax) and figure (fig) explicitly.

Useful when combining multiple plots in the same figure.

Example 1: Using ax = sns...

```python
fig, ax = plt.subplots()
sns.scatterplot(data=tips, x="total_bill", y="tip", ax=ax, hue="sex")
ax.set_title("Scatterplot with Axis Control")
plt.show()
```

Output:

Same scatterplot, but we used ax.set_title() instead of plt.title().

This is useful when you have multiple axes.

Example 2: Multiple Subplots

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.barplot(data=tips, x="day", y="tip", ax=axes[0])
axes[0].set_title("Bar Plot")

sns.boxplot(data=tips, x="day", y="total_bill", ax=axes[1])
axes[1].set_title("Box Plot")

plt.tight_layout()
plt.show()
```

Output:

Two plots side by side, sharing one figure.
👉 Essential for dashboards or reports.

4. Annotations & Labels
✅ Definition:

Adding titles, labels, and context scaling makes plots more informative.

sns.set_context() → Controls overall scaling (paper, notebook, talk, poster).

plt.title(), plt.xlabel(), plt.ylabel() → Add axis labels and titles.

ax.annotate() → Add custom text inside the plot.

Example 1: Titles & Labels

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker")

plt.title("Bill vs Tip by Smoker Status", fontsize=14, fontweight="bold")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

Output:

Title at top, bold and larger.

X and Y axis properly labeled.

Example 2: Context Scaling

```python
sns.set_context("talk")  # Larger font sizes (good for presentations)

sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Boxplot with Talk Context")
plt.show()
```

Output:

Fonts and labels are larger than default (good for slides)

Example 3: Adding Annotations

```python
fig, ax = plt.subplots()
sns.scatterplot(data=tips, x="total_bill", y="tip", ax=ax)


# Annotate one point
ax.annotate("Biggest tip", xy=(50, 10), xytext=(30, 12),
            arrowprops=dict(facecolor="black", shrink=0.05))
plt.show()
```

Output:

Arrow points to a specific data point ((50,10)) with label “Biggest tip”.
👉 Useful for highlighting important insights.

1. sns.FacetGrid()
✅ Definition:

FacetGrid = foundation for multi-plot grids in Seaborn.

It maps a function (like sns.scatterplot) onto subsets of data.

Allows splitting data by rows, columns, or hue

Example: Faceted Scatterplots

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# FacetGrid: split by smoker (rows) and sex (cols)
g = sns.FacetGrid(tips, row="smoker", col="sex", margin_titles=True)
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
plt.show()
```

Output:

A grid of scatterplots:

Rows = Smoker/Non-smoker.

Columns = Male/Female.
👉 Helps compare relationships across categories.

2. sns.PairGrid()
✅ Definition:

Similar to pairplot() but more customizable.

Lets you control different functions for upper, lower, and diagonal plots.

```python
iris = sns.load_dataset("iris")

# PairGrid
g = sns.PairGrid(iris, hue="species")
g.map_upper(sns.scatterplot)    # Upper triangle → scatter
g.map_lower(sns.kdeplot, fill=True)  # Lower triangle → density
g.map_diag(sns.histplot)        # Diagonal → histograms
g.add_legend()
plt.show()
```

Output:

Grid of plots for all variable pairs.

Upper = scatter, Lower = KDE contours, Diagonal = histograms.
👉 More control than pairplot().

3. sns.catplot()
✅ Definition:

High-level function for categorical plots (like bar, box, violin, strip).

Built on FacetGrid → supports faceting with row, col, hue.

Kind can be "bar", "count", "box", "violin", "strip", "swarm".

Example: Catplot with Facets

```python
# Catplot: bar by day, faceted by sex
sns.catplot(data=tips, x="day", y="total_bill", hue="smoker", col="sex", kind="bar")
plt.show()
```

Output:

Two columns (Male/Female).

Bars grouped by smoker/non-smoker.

Each bar shows mean total bill per day.
👉 Easy way to compare categorical variables across groups.

4. sns.relplot()
✅ Definition:

High-level function for relational plots (scatter and line plots).

Built on FacetGrid → supports faceting with row, col, hue, style, size.

Kind can be "scatter" or "line".

Example: Scatter with Facets

```python
# Relplot: total_bill vs tip by time and smoker
sns.relplot(data=tips, x="total_bill", y="tip", hue="day", col="time", style="sex", kind="scatter")
plt.show()
```

Output:

Line plots of brain signal over time.

Different colors for events, line styles for regions.
👉 Perfect for time series comparisons.

**summary:**

| Function      | Purpose                                               | Best For                            |
| ------------- | ----------------------------------------------------- | ----------------------------------- |
| **FacetGrid** | Base grid system (manual mapping of plots)            | Maximum flexibility                 |
| **PairGrid**  | Grid of pairwise variable comparisons                 | Advanced control (upper/lower/diag) |
| **catplot**   | High-level categorical plots (bar, box, violin, etc.) | Comparing categories with facets    |
| **relplot**   | High-level relational plots (scatter, line)           | Multi-dimensional trend analysis    |
